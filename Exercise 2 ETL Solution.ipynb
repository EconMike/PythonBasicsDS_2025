# Data wrangling

```python
df = pd.DataFrame({
        'id': [1,2,3,4,5],
        'Feature1': ['A', 'C', 'E', 'G', 'I'],
        'Feature2': ['B', 'D', 'F', 'H', 'J']})

df2 = pd.DataFrame({
        'id': [1, 2, 6, 7, 8],
        'Feature3': ['K', 'M', 'O', 'Q', 'S'],
        'Feature4': ['L', 'N', 'P', 'R', 'T']})

```

```python
#using the data above do the merging excerises below

#inner join

inner= pd.merge(df,df2, on=["id"])
inner

```

```python
#outer join

outer= pd.merge(df,df2, how="outer", on=["id"])
outer
```

```python
#Left Join

left = pd.merge(df,df2,  how="left", on=["id"])
left 
```

```python
#right join

right= pd.merge(df,df2,  how="right", on=["id"])
right
```
